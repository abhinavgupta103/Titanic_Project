```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
algorithms = ["Support Vector Classifier", "K-Nearest Neighbors", "Random Forest"];
accuracies = [0.85 ,0.78, 0.75]
colors = ["Red", "Green", "Blue"]
x_axis = np.arange(len(accuracies))
tick_locations = []
labels = []
```

```python
for i in range(0, len(accuracies)):
    value = float(x_axis[i])
    tick_location = value+0.4
    tick_locations.append(tick_location)
    plt.bar(x_axis[i], accuracies[i], color=colors[i], align='edge', label=algorithms[i])
    labels.append(algorithms[i])
plt.xticks(tick_locations, algorithms,fontsize =9)
plt.title("Algorithm Accuracies")
plt.xlabel("Algorithm Name", fontsize =12)
plt.ylabel("Algorithm Accuracy", fontsize =12)
plt.legend(labels,bbox_to_anchor=(1, 1))
plt.figure(figsize=(25,15))

plt.savefig('abc.png', dpi=100, bbox_inches='tight')
#plt.show()

```
